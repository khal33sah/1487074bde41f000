```python
from sklearn.preprocessing   import MinMaxScaler 
scaler   =   MinMaxScaler() 
```

```python
#dropping date and lights columns
new_data =data.drop(columns=[ 'date' ,   'lights' ]) 
new_data
```

```python
#normalizing the data
normalised_data   =   pd.DataFrame(scaler.fit_transform(new_data),   columns=new_data.columns) 
```

```python
response_target   =   normalised_data[ 'Appliances' ]
```

```python
#splitting the data into test and train 
from sklearn.model_selection   import  train_test_split 
x_train,   x_test,   y_train,   y_test   =   train_test_split(new_data,   response_target, 
 test_size= 0.3 ,   random_state= 42) 
```

```python
#fitting the data
from sklearn.linear_model import LinearRegression
linear_model   =   LinearRegression()
linear_model.fit(x_train , y_train)
```

```python
predicted_values = linear_model.predict(x_test)
```

```python
#mean absolute error
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 2) 
```

# QUESTION 14

```python
#residual sum of squares
import numpy as np
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2) 
```

```python
#root mean square
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 2) 
```

# QUESTION 17

```python
#checking the lowest and highest features weight
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.5)
ridge_reg.fit(x_train, y_train)
```

```python
def get_weights_data(model, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_data = pd.DataFrame(weights).reset_index()
  weights_data.columns = ['Features', col_name]
  weights_data[col_name].round(3)
  return weights_data
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
```

```python
linear_model_weights = get_weights_data(linear_model, x_train, 'Linear_Model_Weight')
lasso_weights_data = get_weights_data(lasso_reg, x_train, 'Lasso_weight')
```

```python
final_weights = pd.merge(linear_model_weights, lasso_weights_data, on='Features')
final_weights
```
